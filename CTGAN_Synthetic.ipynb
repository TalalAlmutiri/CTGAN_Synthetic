```python
import pandas as pd
from ctgan import CTGAN
```

```python
df_real_data  = pd.read_csv("/content/Housing.csv")
df_real_data.shape
```

```python
# Names of the columns
column_names = list(df_real_data.columns.values)
column_names
```

```python
ctgan = CTGAN(epochs=10)
ctgan.fit(df_real_data, column_names)

# Create synthetic data
synthetic_data = ctgan.sample(500)
```
